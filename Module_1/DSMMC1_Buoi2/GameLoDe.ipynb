```python
import random
from IPython.display import clear_output

class GameLoDe:
    def __init__(self):
        self.tong_tien = 1000000

    def quay_so(self):
        return [random.randint(10, 99) for _ in range(7)]

    def in_danh_sach_giai(self, danh_sach_giai):
        print('-' * 20)
        print("---DANH SÁCH GIẢI---")
        for i, so in enumerate(danh_sach_giai, start=1):
            print(f"Giai {i}: {so}")
        print('-' * 20)

    def nhap_so_lo(self):
        while True:
            so_lo = input("Nhập số lô mà bạn muốn cược (10-99): ")
            if so_lo.isdigit() and 10 <= int(so_lo) <= 99:
                return int(so_lo)
            else:
                print("Nhập sai định dạng, vui lòng nhập lại!")

    def nhap_tien_cuoc(self):
        while True:
            tien_cuoc = input(f"Nhập tiền cược mà bạn muốn cược (tiền cược < {self.tong_tien}): ")
            if tien_cuoc.isdigit() and int(tien_cuoc) < self.tong_tien:
                return int(tien_cuoc)
            else:
                print("Nhập sai định dạng hoặc quá tổng tiền. Nhập lại")

    def game_quay_so(self, so_lo, tien_cuoc):
        danh_sach_giai = self.quay_so()
        print(f"Số lô bạn chơi là: {so_lo}")
        self.in_danh_sach_giai(danh_sach_giai)
        dem_so_lo = danh_sach_giai.count(so_lo)
        if dem_so_lo > 0:
            trung_thuong = (tien_cuoc * dem_so_lo) * 70
            self.tong_tien += trung_thuong
            print(f"Bạn đã trúng {dem_so_lo} nháy!")
            print(f"Số tiền bạn trúng là {trung_thuong}")
            print(f"Tổng tiền bạn có là {self.tong_tien}")
        else:
            print("Bạn đã thua lô!")
            self.tong_tien -= tien_cuoc
            print(f"Tổng tiền bạn còn là {self.tong_tien}")

    def nap_tien(self):
        while True:
            tien_nap = int(input("Nhập số tiền mà bạn muốn nạp: "))
            self.tong_tien += tien_nap
            print(f"Tổng tiền sau khi bạn nạp là {self.tong_tien}")
            nap_tiep = int(input("Bạn có muốn nạp tiếp không? (0: không, 1: có)"))
            if nap_tiep == 0:
                break

    def menu(self):
        print("GAME LÔ ĐỀ HỌC")
        print("1. Chơi lô")
        print("2. Nạp tiền")
        print("3. Thoát game")

    def start(self):
        self.menu()
        while True:
            lua_chon = input("Bạn chọn: ")
            if lua_chon == '1':
                clear_output(wait=True)
                while True:
                    tien_cuoc = self.nhap_tien_cuoc()
                    so_lo = self.nhap_so_lo()
                    self.game_quay_so(so_lo, tien_cuoc)
                    choi_tiep = input("Bạn có muốn chơi tiếp không? (0: không, 1: có)")
                    if choi_tiep == '0':
                        break
                clear_output(wait=True)
                self.start()
                break
            elif lua_chon == '2':
                clear_output()
                self.nap_tien()
                clear_output(wait=True)
                self.start()
                break
            elif lua_chon == '3':
                print("Chào bạn, hẹn gặp lại!")
                break
            else:
                print("Nhập sai lựa chọn, vui lòng nhập lại !")

# Khởi tạo và chạy trò chơi
game = GameLoDe()
game.start()
```
