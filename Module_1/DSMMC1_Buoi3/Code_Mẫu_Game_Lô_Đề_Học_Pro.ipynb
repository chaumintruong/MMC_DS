```python
import random

class LotteryGame:
    def __init__(self):
        self.total_money = 1000000
        self.logged_in = False

    def login(self):
        print("--- CHÀO MỪNG ĐẾN VỚI GAME LÔ ĐỀ ---")
        print("Vui lòng đăng nhập để chơi game")
        while True:
            username = input("Nhập tên đăng nhập: ")
            password = input("Nhập mật khẩu: ")
            if username == "admin" and password == "admin":
                self.logged_in = True
                break
            else:
                print("Đăng nhập không thành công. Vui lòng thử lại.")

    def play_lottery(self):
        if not self.logged_in:
            print("Vui lòng đăng nhập để chơi game.")
            return

        print("--- CHƠI LÔ ĐỀ ---")
        number_of_tickets = int(input("Nhập số lượng vé số bạn muốn chơi: "))
        tickets = self.input_lottery_numbers(number_of_tickets)
        bet_amount = self.place_bet(number_of_tickets)

        results = self.draw_lottery_results()
        self.calculate_winnings(tickets, results, bet_amount)

        self.continue_playing()

    def input_lottery_numbers(self, number_of_tickets):
        tickets = []
        for _ in range(number_of_tickets):
            while True:
                ticket_numbers = input("Nhập số lô (2 chữ số) của bạn, cách nhau bởi dấu phẩy: ")
                if all(len(num.strip()) == 2 and num.strip().isdigit() for num in ticket_numbers.split(',')):
                    tickets.append(ticket_numbers.split(','))
                    break
                else:
                    print("Số lô không hợp lệ. Vui lòng nhập lại.")
        return tickets

    def place_bet(self, number_of_tickets):
        while True:
            bet_amount = int(input("Nhập số tiền cược của bạn: "))
            if bet_amount <= self.total_money:
                self.total_money -= bet_amount
                return bet_amount
            else:
                print("Số tiền cược vượt quá số dư của bạn. Vui lòng nhập lại.")

    def draw_lottery_results(self):
        results = {}
        prizes = ["GIẢI NHẤT", "GIẢI NHÌ", "GIẢI BA", "GIẢI TƯ", "GIẢI NĂM", "GIẢI SÁU", "GIẢI BẢY"]
        for prize in prizes:
            result = ''.join(str(random.randint(0, 9)) for _ in range(5))
            results[prize] = result
        self.print_lottery_results(results)
        return results

    def print_lottery_results(self, results):
        print("--- KẾT QUẢ QUAY SỐ ---")
        for prize, result in results.items():
            print(f"{prize}: {result}")

    def calculate_winnings(self, tickets, results, bet_amount):
        winnings = 0
        for ticket in tickets:
            for prize, result in results.items():
                if ticket == result[-2:]:
                    winnings += bet_amount * 70
        self.total_money += winnings
        print(f"Bạn đã trúng {len(tickets)} nháy!")
        print(f"Số tiền bạn trúng là: {winnings}")
        print(f"Tổng số tiền bạn có là: {self.total_money}")

    def continue_playing(self):
        choice = input("Bạn có muốn chơi tiếp không? (Y/N): ").lower()
        if choice == 'y':
            self.play_lottery()
        else:
            print("Cảm ơn bạn đã tham gia. Hẹn gặp lại!")

def main():
    game = LotteryGame()
    game.login()
    game.play_lottery()

if __name__ == "__main__":
    main()
```
